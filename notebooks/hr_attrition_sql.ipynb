## convert the ReviewDate column in the Performance table to a DATE format and updates the table with a new FormattedDate column.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('//content/drive/MyDrive/portfolio/Project/HR_database.db')

df = pd.read_sql_query("SELECT * FROM Performance", conn)

df['FormattedDate'] = pd.to_datetime(df['ReviewDate'], format='%m/%d/%Y', errors='coerce')

df['FormattedDate'] = df['FormattedDate'].dt.strftime('%Y-%m-%d')

cursor = conn.cursor()
cursor.execute("PRAGMA table_info(Performance)")
columns = cursor.fetchall()

if not any(col[1] == 'FormattedDate' for col in columns):
    cursor.execute('''
        ALTER TABLE Performance
        ADD COLUMN FormattedDate DATE
    ''')

for i, row in df.iterrows():
    cursor.execute('''
        UPDATE Performance
        SET FormattedDate = ?
        WHERE PerformanceID = ?
    ''', (row['FormattedDate'], row['PerformanceID']))

conn.commit()
conn.close()
```

**Is there a link between job satisfaction and attrition in the most recent year 2022?**

```python
%%sql
with recentreviews as (
  select * from Performance
  where strftime('%Y',FormattedDate) = '2022'
) ,
SatisfactionAttrition as (
  select p.JobSatisfaction,e.Attrition
  from recentreviews p
  join Employee e on p.EmployeeID = e.EmployeeID
)
select JobSatisfaction,avg(case when Attrition = 'Yes' then 1 else 0 end) as AttritionRate
from SatisfactionAttrition
group by JobSatisfaction
```

Employees with the lowest satisfaction (level 1) have the highest attrition rate (26.6%), while those with higher satisfaction levels have lower attrition rates. This indicates dissatisfaction plays a key role in turnover, potentially due to feeling undervalued or lacking growth opportunities.

To address this, the company could focus on improving the work environment, offering growth opportunities, and engaging dissatisfied employees (level 1) first, as they are at the highest risk of leaving.

**Is there a link between Relationship satisfaction and attrition in the most recent year 2022?**

```python
%%sql
with recentreviews as (
  select * from Performance
  where strftime('%Y',FormattedDate) = '2022'
) ,
SatisfactionAttrition as (
  select p.RelationshipSatisfaction,e.Attrition
  from recentreviews p
  join Employee e on p.EmployeeID = e.EmployeeID
)
select RelationshipSatisfaction,avg(case when Attrition = 'Yes' then 1 else 0 end) as AttritionRate
from SatisfactionAttrition
group by RelationshipSatisfaction
```

Employees with moderate relationship satisfaction (level 3) have the highest attrition rate (25.1%), while those with very low (level 1) and very high (level 5) satisfaction show slightly lower but still notable attrition rates at 22.4% and 19.8%, respectively. To address this, the company could foster a more connected and supportive work culture, encourage team collaboration, and provide conflict resolution resources to help employees build stronger workplace relationships.

**Is there a link between Work-Life Balance and attrition in the most recent year 2022?**

```python
%%sql
with recentreviews as (
  select * from Performance
  where strftime('%Y',FormattedDate) = '2022'
) ,
worklifeAttrition as (
  select p.WorkLifeBalance,e.Attrition
  from recentreviews p
  join Employee e on p.EmployeeID = e.EmployeeID
)
select WorkLifeBalance,avg(case when Attrition = 'Yes' then 1 else 0 end) as AttritionRate
from worklifeAttrition
group by WorkLifeBalance
```

Building upon our previous study of correlations between these factors and employee outcomes, I developed a simple classification model to categorize employees into High Risk and Low-Middle Risk groups.

**Categorizing**

```python
df = pd.read_sql('''
    SELECT *
    FROM Employee
    JOIN Performance ON Employee.EmployeeID = Performance.EmployeeID
    WHERE (strftime("%Y", FormattedDate) = "2022")
    AND Attrition = 'No'
''', conn)

def classify_risk(row):
    risk_score = 0

    if 18 <= row['Age'] < 30:
        risk_score += 1

    if row['JobSatisfaction'] < 2:
        risk_score += 1

    if row['RelationshipSatisfaction'] < 2:
        risk_score += 1

    if row['WorkLifeBalance'] < 2:
        risk_score += 1

    if risk_score >= 2:
        return 'High Risk'
    elif risk_score == 1:
        return 'Middle Risk'
    else:
        return 'Low Risk'

df['RiskCategory'] = df.apply(classify_risk, axis=1)

display(df[['EmployeeID', 'FirstName', 'LastName', 'Age', 'JobSatisfaction', 'RiskCategory']].head())

```

## Performance

The Performance section examines employee ratings by managers, focusing on average ratings across different demographics and work history. This analysis highlights potential trends in performance scores among different demographics. Additionally, the correlation between job satisfaction and average ratings was explored to understand whether higher satisfaction aligns with better performance.

**By Gender**

```python
%%sql
select Gender,avg(ManagerRating) as  AverageManagerRating
from performance join Employee
on performance.EmployeeID = Employee.EmployeeID
where Attrition = 'No' And strftime('%Y',FormattedDate) = '2022'
group by Gender
```

The average Manager Rating varies slightly across gender groups. Male employees have the highest average rating (3.5), while Non-Binary employees have the lowest (3.30). Female employees (3.44) and those who Prefer Not to Say (3.42) fall in between. This suggests potential differences in performance evaluations, which may be worth exploring further to ensure fairness and consistency.

**By Age Group**

```python
%%sql
SELECT CASE
   WHEN Age BETWEEN 18 AND 30 THEN 'Young'
   WHEN Age BETWEEN 31 AND 45 THEN 'Middle-Aged'
   WHEN Age BETWEEN 46 AND 59 THEN 'Older' END AS AgeGroup , avg(ManagerRating) as ManagerRating
from Employee join performance on employee.EmployeeID = performance.EmployeeID
where Attrition = 'No' And (strftime('%Y',FormattedDate) = '2022')
group by AgeGroup
```

The average Manager Rating varies by age group. Young employees have the highest rating (3.48), followed by middle-aged employees (3.46). Older employees have the lowest average rating (3.31). This could suggest different perceptions of performance across age groups, which may require further analysis.

**By Department**

```python
%%sql
select department,avg(ManagerRating) as  AverageManagerRating
from performance join Employee
on performance.EmployeeID = Employee.EmployeeID
where Attrition = 'No' And strftime('%Y',FormattedDate) = '2022'
group by department
```

The Performance analysis shows how employees are rated by their managers across different departments. HR has the highest average rating (3.66), followed by Sales (3.46) and Technology (3.44).These differences could be due to factors like job roles, expectations, or work dynamics.

**By Job Role**

```python
%%sql
select JobRole,avg(ManagerRating) as  AverageManagerRating
from performance join Employee
on performance.EmployeeID = Employee.EmployeeID
where Attrition = 'No' And strftime('%Y',FormattedDate) = '2022'
group by JobRole order by AverageManagerRating desc
```

Manager ratings vary across different job roles. Recruiters have the highest average rating (3.9), followed by HR Business Partners (3.75) and Data Scientists (3.61). On the other hand, Sales Representatives have the lowest rating (3.23), with Engineering Managers (3.29) and Software Engineers (3.39) also on the lower end.

The differences might be due to job expectations, how performance is measured, or the nature of each role. It could be useful to look into whether job satisfaction, workload, or work environment affect these ratings.

**By Years At Company**

```python
%%sql
SELECT CASE
        WHEN YearsAtCompany BETWEEN 0 AND 1 THEN '0-1 years'
        WHEN YearsAtCompany BETWEEN 2 AND 5 THEN '2-5 years'
        WHEN YearsAtCompany BETWEEN 6 AND 10 THEN '6-10 years' END AS YearsAtCompanyGroup , avg(ManagerRating) as ManagerRating
FROM Employee join performance on employee.EmployeeID = performance.EmployeeID
where attrition = 'No' and strftime('%Y',FormattedDate) = '2022' GROUP BY YearsAtCompanyGroup
```

Employees in their first year have the highest average manager rating (3.62), while those with 2-5 years receive the lowest (3.41). Ratings then improve slightly for employees with 6-10 years (3.48). This suggests that new employees may start off strong, possibly due to initial enthusiasm or manager expectations. However, ratings drop after a few years, which could be due to increased job demands or shifting expectations. The slight improvement after six years might indicate that experience and long-term stability lead to better performance evaluations.
